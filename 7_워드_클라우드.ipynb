```python
text = open('/content/drive/MyDrive/KDT/2. 데이터분석/alice.txt').read()
text
```

```python
from wordcloud import WordCloud
```

```python
from konlpy.corpus import kolaw
```

```python
text = kolaw.open('constitution.txt').read()
```

```python
text
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
noun_text = okt.nouns(text)
print(noun_text)
```

```python
noun_text.sort(key=lambda x: len(x))
print(noun_text)
```

```python
stop_words = ['함', '것', '제', '정', '바', '그']
noun_text = [each_word for each_word in noun_text if each_word not in stop_words]
print(noun_text)
```

```python
noun_text = [each_word for each_word in noun_text if len(each_word) > 1]
print(noun_text)
```

```python
from collections import Counter
```

```python
count = Counter(noun_text)
print(count)
```

```python
data = count.most_common(100)
data = dict(data)
print(len(data))
print(data)
```

```python
# 딕셔너리에 저장된 데이터를 워드 클라우드로 생성하려면 generate_from_frequencies()를 사용
wordcloud = WordCloud(max_words=100,
                      font_path='/usr/share/fonts/truetype/nanum/NanumPen.ttf'
                      ).generate_from_frequencies(data)
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
mask = np.array(Image.open('/content/drive/MyDrive/KDT/2. 데이터분석/korea_mask.jpg'))
```

```python
# 딕셔너리에 저장된 데이터를 워드 클라우드로 생성하려면 generate_from_frequencies()를 사용
wordcloud = WordCloud(max_words=100,
                      font_path='/usr/share/fonts/truetype/nanum/NanumPen.ttf',
                      background_color='ivory',
                      mask=mask
                      ).generate_from_frequencies(data)
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
